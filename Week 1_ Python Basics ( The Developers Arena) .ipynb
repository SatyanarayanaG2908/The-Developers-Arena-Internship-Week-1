## Exercise 2: Variables and Data Types

```python
# variables_practice.py

# String variables
first_name = "John"
last_name = "Smith"
favorite_quote = "To be or not to be"

# Number variables
age = 25
height = 5.9
temperature = -3.5

# Boolean variables
is_student = True
has_drivers_license = False

# Print all variables
print("Name:", first_name, last_name)
print("Age:", age)
print("Height:", height, "feet")
print("Temperature:", temperature, "degrees")
print("Student?", is_student)
print("Has license?", has_drivers_license)

# Check types
print("\nVariable Types:")
print(f"first_name is a {type(first_name)}")
print(f"age is a {type(age)}")
print(f"height is a {type(height)}")
print(f"is_student is a {type(is_student)}")
```

## Exercise 3: Simple Calculator
### Goal: Build a calculator that performs basic math operations

```python
# simple_calculator.py

print("=== Simple Calculator ===")
print()

# Get numbers from user
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))

# Perform calculations
addition = num1 + num2
subtraction = num1 - num2
multiplication = num1 * num2
division = num1 / num2
floor_division = num1 // num2
modulus = num1 % num2

try:
    exponentiation = num1 ** num2
except OverflowError:
    exponentiation = "Result too large to display"

# Display results
print("\n=== Results ===")
print(f"{num1} + {num2} = {addition}")
print(f"{num1} - {num2} = {subtraction}")
print(f"{num1} * {num2} = {multiplication}")
print(f"{num1} / {num2} = {division}")
print(f"{num1} // {num2} = {floor_division}")
print(f"{num1} % {num2} = {modulus}")
print(f"{num1} ** {num2} = {exponentiation}")
```

## Exercise 5: Fixing Syntax Errors
### Goal: Practice identifying and fixing common errors

```python
# error_practice.py
# Try to find and fix the errors in this code!

# Error 1: Missing closing quote
# print("Hello, World!)
# FIX:
print("Hello, World!")

# Error 2: Variable name starts with number
# 1st_name = "John"
# FIX:
first_name = "John"

# Error 3: Wrong type for operation
# age = "25"
# next_year_age = age + 1
# FIX:
age = 25
next_year_age = age + 1
print(f"Next year you'll be {next_year_age}")

# Error 4: Undefined variable
# print(user_name)
# FIX:
user_name = "Alice"
print(user_name)

# Error 5: Mixing quotes
# message = "Hello, it's me"
# FIX:
message = "Hello, it's me"
# OR
message = 'Hello, it\'s me'  # Using escape character
print(message)
```

# üéØ Week 1 Project: Personal Information Manager
### Goal: Create a comprehensive program that collects and displays personal information with formatted output.

```python
# personal_info_manager.py

"""
Personal Information Manager
Week 1 Project - Python Basics

This program collects personal information from the user
and displays it in a formatted, organized manner.
"""

# Display welcome banner
print("=" * 50)
print("    PERSONAL INFORMATION MANAGER")
print("=" * 50)
print()

# Collect user information
print("Please enter your information:")
print("-" * 50)

name = input("Full Name: ")
age = int(input("Age: "))
city = input("City: ")
country = input("Country: ")
email = input("Email: ")
phone = input("Phone Number: ")

# Collect hobbies (as a single string, comma-separated)
hobbies = input("Hobbies (separate with commas): ")

# Optional: Favorite things
favorite_color = input("Favorite Color: ")
favorite_food = input("Favorite Food: ")
favorite_book = input("Favorite Book: ")

# Calculate birth year (approximate)
current_year = 2025
birth_year = current_year - age

# Clear screen effect (print empty lines)
print("\n" * 2)

# Display formatted output
print("=" * 50)
print("       YOUR PERSONAL INFORMATION")
print("=" * 50)
print()

# Personal Details Section
print("üìã PERSONAL DETAILS")
print("-" * 50)
print(f"Name:              {name}")
print(f"Age:               {age} years old")
print(f"Birth Year:        ~{birth_year}")
print(f"Location:          {city}, {country}")
print()

# Contact Information Section
print("üìû CONTACT INFORMATION")
print("-" * 50)
print(f"Email:             {email}")
print(f"Phone:             {phone}")
print()

# Interests Section
print("üé® INTERESTS & HOBBIES")
print("-" * 50)
print(f"Hobbies:           {hobbies}")
print()

# Favorites Section
print("‚≠ê FAVORITES")
print("-" * 50)
print(f"Color:             {favorite_color}")
print(f"Food:              {favorite_food}")
print(f"Book:              {favorite_book}")
print()

# Fun Facts Section
print("üí° FUN FACTS")
print("-" * 50)

# Count number of hobbies
hobby_list = hobbies.split(",")
hobby_count = len(hobby_list)
print(f"‚Ä¢ You have {hobby_count} hobbies!")

# Calculate days lived (approximate)
days_lived = age * 365
print(f"‚Ä¢ You've lived approximately {days_lived:,} days!")

# Calculate age in months
months_lived = age * 12
print(f"‚Ä¢ That's about {months_lived} months!")

# Name length
name_length = len(name)
print(f"‚Ä¢ Your name has {name_length} characters (including spaces)!")

print()

# Closing message
print("=" * 50)
print(f"Thank you, {name}!")
print("Your information has been recorded successfully.")
print("=" * 50)

# Save option (just a message, no actual file saving yet)
print()
save_option = input("Would you like to save this information? (yes/no): ")
if save_option.lower() == "yes":
    print("‚úì Information saved! (Note: Actual saving will be covered in Week 3)")
else:
    print("Information not saved.")

print()
print("Program finished. Have a great day!")
```
